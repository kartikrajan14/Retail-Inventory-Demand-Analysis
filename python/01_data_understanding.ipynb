```python
import pandas as pd

df = pd.read_csv("retail_store_inventory.csv")
df.head()
```

```python
df.describe()
```

```python
df.loc[df['Demand Forecast'] < 0, 'Demand Forecast'] = 0
```

```python
df['Demand Forecast'].describe()
```

```python
df['Stockout_Risk'] = np.where(
    (df['Inventory Level'] < 100) & (df['Units Sold'] > 150),
    1,
    0
)
```

```python
df['Stockout_Risk'].value_counts()
```

```python
stockout_by_category = (
    df.groupby('Category')['Stockout_Risk']
      .mean()
      .sort_values(ascending=False)
)

stockout_by_category
```

```python
df['Forecast_Error'] = df['Demand Forecast'] - df['Units Sold']
```

```python
df['Abs_Forecast_Error'] = df['Forecast_Error'].abs()
```

```python
df['Abs_Forecast_Error'].mean()
```

```python
forecast_error_by_category = (
    df.groupby('Category')['Abs_Forecast_Error']
      .mean()
      .sort_values(ascending=False)
)

forecast_error_by_category
```

```python
df['Discount_Flag'] = np.where(df['Discount'] > 0, 1, 0)
```

```python
discount_sales_comparison = (
    df.groupby('Discount_Flag')['Units Sold']
      .mean()
)

discount_sales_comparison
```

```python
region_sales = (
    df.groupby('Region')['Units Sold']
      .sum()
      .sort_values(ascending=False)
)

region_sales
```
